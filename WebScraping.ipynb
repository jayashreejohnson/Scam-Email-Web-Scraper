```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from urllib.parse import urljoin

base_url = "https://www.419scam.org/emails/"

response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'html.parser')

email_list_links = []
for link in soup.find_all('a', href=True):
    href = link['href']
    if href.startswith("20"):
        full_url = urljoin(base_url, href.rstrip('/'))
        email_list_links.append(full_url)

email_list_links = email_list_links[:2]

email_data = []

def extract_date_from_url(url):
    """ Extracts date from a properly formatted email URL """
    parts = url.replace(base_url, "").split("/")
    if len(parts) >= 2 and parts[0].startswith("20"):
        return f"{parts[0]}/{parts[1]}"
    return "Invalid"

for list_page in email_list_links:
    response = requests.get(list_page)
    soup = BeautifulSoup(response.text, 'html.parser')

    for email_link in soup.find_all('a', href=True):
        email_href = email_link['href']
        if email_href.endswith(".htm"):
            email_url = urljoin(list_page, email_href)
            link_text = email_link.text.strip()
            date_part = extract_date_from_url(email_url)

            if date_part != "Invalid":
                email_data.append([link_text, email_url, date_part])

    time.sleep(1)

df = pd.DataFrame(email_data, columns=["Link Text", "URL", "Date"])
df = df[~df["Link Text"].str.match(r"^\d{4}-\d{2}$|^\d{2}$")]

df.reset_index(drop=True, inplace=True)

df
```
